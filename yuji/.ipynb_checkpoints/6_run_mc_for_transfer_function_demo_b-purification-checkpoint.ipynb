# Run MCs w/ observation matrix to get a transfer function

```python
mcnums = 1024
```

```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
PREFIX = 'observed'
nside = 64
```

```python
Npix = hp.nside2npix(nside)
lmax = 3 * nside - 1
sigmab = hp.nside2resol(nside)
fwhm = (8*np.log(2))**0.5 * sigmab
print('nside:{:}, Npix:{:}, res:{:.1f} deg, sigmab:{:.1f} deg, fwhm:{:.1f} deg'.
      format(nside, Npix, np.rad2deg(hp.nside2resol(nside)), np.rad2deg(sigmab), np.rad2deg(fwhm)))
```

```python
prefix = 'scan_masked'
mask = np.array(hp.read_map('{:}_nside{:03d}.fits'.format(prefix,nside)), dtype=bool)
#hp.mollview(mask, title='Used mask, ring')
unseen = ~mask
nums = np.sum(mask)
fsky = nums / mask.shape[0]
fsky
```

```python
prefix = 'scan_masked_apodization'
Z = apodization = np.array(hp.read_map('{:}_nside{:03d}.fits'.format(prefix,nside)))
#hp.mollview(np.ma.masked_array(apodization, mask=~mask, fill_value=hp.UNSEEN), title='Used apodization, ring')
```

## prepare camb

```python
import camb
```

```python
%%time
pars = camb.CAMBparams()
pars.InitPower.set_params(As=2e-9, ns=0.965, r=0.2)
pars.set_for_lmax(4000, lens_potential_accuracy=1)
pars.set_accuracy(AccuracyBoost=1)
pars.set_cosmology(H0=67.5, ombh2=0.022, omch2=0.122, mnu=0.06, omk=0, tau=0.06)
pars.WantTensors = True
pars.max_l_tensor = 1600
pars.max_eta_k_tensor = 3200
results = camb.get_results(pars)
powers =results.get_cmb_power_spectra(pars, CMB_unit='muK')
print(powers.keys())
```

```python
def dl2cl(l,dl):
    return dl/l/(l+1)*2*np.pi

def cl2dl(l,cl):
    return l*(l+1)*cl/2/np.pi

def get_cls(key, teb='teb'):
    #total
    #unlensed_scalar
    #unlensed_total
    #lensed_scalar
    #tensor
    #ens_potential
    Dls = powers[key]
    ell = np.arange(Dls.shape[0])
    ell[0] = 1.
    cltt, clee, clbb, clte = dl2cl(ell, Dls.T)
    if teb == 'te':
        clbb = np.zeros_like(cltt)
    elif teb == 'tb':
        clte = clee = np.zeros_like(cltt)

    cltb = cleb = np.zeros_like(cltt)
    # a list of clâ€™s (TT, EE, BB, TE, EB, TB for polarized input map)
    return cltt,clee,clbb,clte,cleb,cltb
```

## prepare purification

```python
import scipy
```

```python
%%time
idx_n2r = hp.ring2nest(nside,np.arange(Npix)[mask])
mask2 = np.hstack((idx_n2r, idx_n2r + Npix))
obs_matrix = scipy.sparse.load_npz('toast_telescope_all_time_all_obs_matrix.npz')[Npix:,Npix:][np.ix_(mask2,mask2)]
```

```python
%%time
from scipy.sparse import diags
#Z = apodization * nums / np.sum(apodization)
ZZ = diags(np.hstack((Z[mask],Z[mask])))
obs_matrix_apo = ZZ @ obs_matrix
```

```python
%%time
f = np.load('{:}_eigen_spectrum_nside{:03d}.npz'.format(PREFIX,nside))
eigvals = f['eigvals']
norm_eigvecs = f['norm_eigvecs']
```

```python
#lens = np.searchsorted(eigvals > 0.98, True)
lens = 2 * nums //4
2 * nums, lens
```

```python
%%time
v = norm_eigvecs[:,-lens:]
Pib = Pi = np.tensordot(v,v,axes=((1,),(1,)))
```

## prepare nmt

```python
import pymaster as nmt

def maps2pseudospectra(iqu_map, apod, has_apodized=False, do_pure=False):
    if has_apodized:
        amask = apod > 0.
        tmp = np.zeros_like(iqu_map)
        tmp[:,amask] = iqu_map[:,amask]/apod[amask]
        I,Q,U = tmp
    else:
        I,Q,U = iqu_map
        
    f0 = nmt.NmtField(apod, [I,])
    (cltt,) = nmt.compute_coupled_cell(f0, f0)

    f2 = nmt.NmtField(apod, [Q, U], purify_e=False, purify_b=do_pure)
    clee,cleb,clbe,clbb = nmt.compute_coupled_cell(f2, f2)
    
    clte,cltb = nmt.compute_coupled_cell(f0, f2)

    return cltt,clee,clbb,clte,0.5*(cleb+clbe),cltb,np.arange(len(cltt))
```

## run MCs

```python
def fill_pcls(res, cls):
    res[...] = cls[:3]
```

```python
%%time
dicCls = {}
dicCLs = {}
np.random.seed(0)
for teb, flavor in zip(('te', 'tb', 'teb'),('ttee', 'ttbb', 'tteebb')):
    # LCDM; 
    key = 'total'
    Cls = get_cls(key, teb=teb) # sim input w/o beam
    l = ls = np.arange(len(Cls[0]))
    Dls = l*(l+1)/2./np.pi*Cls
    CLs = Cls * np.exp(-l*(l+1)*sigmab**2) # w/ beam for reference
    dicCls[flavor] = Cls
    dicCLs[flavor] = CLs
    DLs = cl2dl(l,CLs)

    # at idx=1, 0:input, 1: apodized anafast, 2: KS purification, 3: KS no purification, 4: B purification
    mcres = np.zeros(shape=(mcnums, 5, 3, lmax+1))
    
    print(flavor, end=': ')
    for i in range(mcnums):
        if i % 5 == 0:
            print(i, end=',')

        map0 = hp.synfast(Cls, nside, sigma=sigmab, pol=True, new=True) # beam convolved when simulated here
        m = np.zeros(shape=(3,Npix))
        Rm = np.zeros(shape=(3,Npix))
        pRm = np.zeros(shape=(3,Npix))
        
        m[:,mask] = map0[:,mask]
        Rm[1:,mask] = (obs_matrix_apo @ m[1:,mask].ravel()).reshape((2,-1))
        pRm[1:,mask] = (Pib @ Rm[1:,mask].ravel()).reshape((2,-1))

        # input
        cls = hp.anafast(map0, pol=True)
        fill_pcls(mcres[i][0], cls)
    
        # masked
        cls = hp.anafast(Rm, pol=True)
        fill_pcls(mcres[i][1], cls)
    
        # KS purif
        cls = maps2pseudospectra(Rm, apodization, has_apodized=True, do_pure=True)
        fill_pcls(mcres[i][2], cls)
        
        # no KS purif
        cls = maps2pseudospectra(Rm, apodization, has_apodized=True, do_pure=False)
        fill_pcls(mcres[i][3], cls)

        # B-purification
        cls = hp.anafast(pRm, pol=True)
        fill_pcls(mcres[i][4], cls)

    print(mcnums-1, 'done')

    avg = np.abs(np.mean(mcres, axis=0))
    std = np.std(mcres, axis=0)

    for _k,_v in zip(('res', 'avg', 'std'),(mcres, avg, std)):
        fn = 'sim_{:}_{:}_mc{:04d}'.format(flavor, _k, mcnums)
        np.save(fn, _v)
```

## plot results

```python
flavor = 'tteebb'
avg = np.load('sim_{:}_avg_mc{:04d}.npy'.format(flavor,mcnums))
std = np.load('sim_{:}_std_mc{:04d}.npy'.format(flavor,mcnums))

xmin = 5.5
xmax = 2 * nside # lmax
ymax= 3e-3
ymin= 1e-10
pminmax = (ymin,ymax)

l = np.arange(lmax+1)
fig,ax = plt.subplots(2,2, figsize=(12,12))
for ii,(aflavor,(vmin,vmax)) in enumerate(zip(('EE','BB'),(pminmax,pminmax))):
    for i,y in enumerate((avg, std)):
        plt.axes(ax[ii][i])
        plt.plot(l, dicCLs[flavor][ii+1][:lmax+1], 'k--')
        plt.plot(l, y[0][ii+1], 'ro', alpha=0.25, label='input')
        plt.xscale('log')
        plt.xlim(xmin, xmax)
        plt.ylim(vmin, vmax)
        plt.xlabel(r"Multipole Moment, $\ell$")
        plt.ylabel(r"$C_\ell$")
        plt.yscale('log')
        if i == 0:
            plt.title('T+E+B Observed {:} Mean (#:{:})'.format(aflavor,mcnums))
        else:
            plt.title('T+E+B Observed {:} STD (#:{:})'.format(aflavor,mcnums))

        plt.plot(l, y[1][ii+1], 'r-', label='anafast')
        plt.plot(l, y[2][ii+1], 'b-', label='KS B-purif')
        plt.plot(l, y[3][ii+1], 'g--', label='KS no purif')
        plt.plot(l, y[4][ii+1], 'c-', label='B-Purified')
        plt.legend()
```
