# combine calculated sparse analytical covariance

```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
from matplotlib import cm as cm
```

```python
PREFIX = 'scan_masked'
```

```python
nside = 64
njobs_off = 48
njobs_diag = 48
```

```python
Npix = hp.nside2npix(nside)
mask = hp.read_map('{:}_nside{:03}.fits'.format(PREFIX,nside))
npix = int(np.sum(mask))
nums = 2 * npix
#hp.mollview(mask)
npix, Npix
```

## combine, save and plot

```python
from scipy.sparse import csc_matrix as smatrix
```

```python
%%time
for key in ('only_bb','only_ee'):
    cov_off = smatrix((nums,nums),dtype=np.float64)
    cov_diag = smatrix((nums,nums),dtype=np.float64)
    
    for ijobs in range(njobs_off):
        print('off: {:}/{:}, '.format(ijobs+1,njobs_off), end='')
        fn = './nside{:03d}/{:}_ana_cov_{:}_off_{:03d}-{:03d}_nisde{:03d}.npz'.format(nside,PREFIX,key,ijobs+1,njobs_off,nside)
        vec = np.load(fn)['res'].ravel()
        cov_off += smatrix((vec[2::3], (np.int_(vec[0::3]), np.int_(vec[1::3]))), shape=(nums,nums), dtype=np.float64)

    for ijobs in range(njobs_diag):   
        print('diag: {:}/{:}, '.format(ijobs+1,njobs_diag), end='')
        fn = './nside{:03d}/{:}_ana_cov_{:}_diag_{:03d}-{:03d}_nisde{:03d}.npz'.format(nside,PREFIX,key,ijobs+1,njobs_diag,nside)
        vec = np.load(fn)['res'].ravel()
        cov_diag += smatrix((vec[2::3], (np.int_(vec[0::3]), np.int_(vec[1::3]))), shape=(nums,nums), dtype=np.float64)

    print(key, 'is done;')

#    np.save('{:}_ana_cov_{:}_nisde{:03d}.npy'.format(PREFIX, key, nside), (cov_off + cov_diag).todense()) # for lower triangle
    np.save('{:}_ana_cov_{:}_nisde{:03d}.npy'.format(PREFIX, key, nside), (cov_off + cov_off.T + cov_diag).todense()) # for full
```

```python
%%time
fig,ax = plt.subplots(1, 2, figsize=(14,7))
vmax = 1./5
for idx,key in enumerate(('only_bb','only_ee')):
    cov = np.load('{:}_ana_cov_{:}_nisde{:03d}.npy'.format(PREFIX, key, nside))
    med = np.median(cov.diagonal())

    plt.axes(ax[idx])
    plt.title('{:} {:} cov, ana, nside={:d}'.format(PREFIX, key, nside))
    im = plt.imshow(cov/med, 'bwr', interpolation=None, vmax=vmax, vmin=-vmax)
    cbr = plt.colorbar(im,fraction=0.046, pad=0.04)
    cbr.set_label('normalized correlation')
plt.savefig('{:}_normalized_corr_nside{:03d}.png'.format(PREFIX,nside))
```
