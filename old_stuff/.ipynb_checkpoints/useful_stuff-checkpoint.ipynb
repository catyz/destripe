```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
m = rng.standard_normal(2**9) #random values for pixelized map, m
```

```python
n_pass = 10
P = np.zeros((len(m)*n_pass, len(m)))

for row in P:
    index = rng.integers(0, len(m))
    row[index] = 1
```

```python
n = rng.normal(scale=1, size = P.shape[0]) #white noise
```

```python
x = np.linspace(0, 4*np.pi, P.shape[0])
n_corr = 3*np.sin(2*x) + 3*np.sqrt(2*x)
```

```python
y = P@m + n_corr + n
# y = n_corr
```

```python
plt.plot(y)
# plt.plot(P@m)
plt.grid()
```

```python
C_n = np.eye(P.shape[0]) #single detector so doesn't matter anyway
```

```python
#Baselines matrix F

n_baseline = 2**7

F = np.zeros((n_baseline, len(y)))

assert len(y) % n_baseline == 0

cutoff = len(y) // n_baseline
offset = 0 
for i in range(n_baseline):
    F[i][offset:offset+cutoff] = 1
    offset+=cutoff
    
F = F.T
```

```python
#destriping. C_n = C_n ^-1 in this simple case 
Z = np.eye(len(y)) - P @ np.linalg.inv(P.T @ C_n @ P) @ P.T @ C_n

#Ax = b
A = F.T @ C_n @ Z @ F
b = F.T @ C_n @ Z @ y
```

```python
#Solution for amplitudes
a = np.linalg.solve(A, b)
```

```python
plt.plot(n_corr, label='input n_corr')
plt.plot(F @ a, label='F * a')
plt.grid()
plt.legend()
```

```python
m_solver = np.linalg.inv(P.T @ C_n @ P) @ P.T @ C_n 

m_destripe = m_solver @ (y - F @ a) #Destriping solution
m_bin = m_solver @ y #Binned map solution

#remove monopole
# m_destripe -= np.mean(m_destripe)
# m_bin -= np.mean(m_bin)
```

```python
plt.plot(m_bin, label='bin')
plt.plot(m_destripe, label='destripe')
plt.plot(m, label='input')
plt.grid()
plt.legend()
```

```python
plt.plot(m_bin - m, label='bin')
plt.plot(m_destripe - m, label='destripe')
plt.grid()
plt.legend()
np.mean(m_destripe-m)
```
