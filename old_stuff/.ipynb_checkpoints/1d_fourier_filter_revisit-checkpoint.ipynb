```python
import numpy as np 
from scipy.linalg import dft
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 6*np.pi, 2**10)
N = len(x)

m_h = 0.3*np.sin(10*x) # High freq component 
m_l = np.sin(x) # Low freq component

m = m_h + m_l
```

```python
plt.plot(m, label='Total')
plt.plot(m_h, label='High freq')
plt.plot(m_l, label='Low freq')
plt.legend()
```

```python
def DFT_matrix(N):
    i, j = np.meshgrid(np.arange(N), np.arange(N))
    omega = np.exp( - 2 * np.pi * 1j / N )
    W = omega ** (i * j ) #/ np.sqrt(N)
    return W

# x = np.zeros(N, N) # x is any input data with those dimensions
# W = DFT_matrix(N)
# dft_of_x = W.dot(x).dot(W)
```

```python
D = DFT_matrix(N)
D_inv = np.linalg.inv(D)
D.shape
```

```python
fft = D @ m
plt.plot(fft)
```

```python
fft[1000:] = 0
fft[:-1000] = 0
```

```python
plt.plot(fft)
len(fft)
```

```python
plt.plot(D_inv @ fft)
# plt.plot(m_h)
```

```python
#now lets shorten
short = fft[fft!=0]
len(short)
```
